3. Check whether all the zones generate the same amount of sales.

Hypothesis
H0: All Zones generate same amount of sales
H1: All zones generate different amount of sales

```python
fvalue,pvalue=stats.f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
fcriteria=stats.f.ppf(1-0.05,3,112)
```

```python
print('F Statistics=',fvalue)
print('F criteria=',fcriteria)
print('pvalue=',pvalue)
print('significance level=0.05')
```

```python
if fvalue>fcriteria and pvalue<0.05:
   print('Reject H0: All Zones generate same amount of sales')
else:
   print('Reject H1: All zones generate different amount of sales')
```
