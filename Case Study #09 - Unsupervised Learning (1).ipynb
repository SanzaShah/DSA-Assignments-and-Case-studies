### Standard Scaling

```python
data1=data
cols= data1.columns
from sklearn.preprocessing import StandardScaler
st_scale= StandardScaler()
data1=st_scale.fit_transform(data1)
```

```python
X=pd.DataFrame(data1,columns=cols)
X.head()
```

### Unsupervised learning

#### K-means Clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
wcss=[]
silhouette_coefficients = []
for i in range(2,11):
    kmeans =KMeans(n_clusters = i, init = 'k-means++',random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_coefficients.append(score)
    print (kmeans.inertia_)
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
plt.plot(range(2, 11), silhouette_coefficients,'-x')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

Optimal number of clusters = 3

```python
#kmeans
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
type(X)
```

```python
y_kmeans
```

```python
data['cluster']=kmeans.labels_
data.head()
```

```python
import seaborn as sns
plt.figure(figsize=(10,20))
plt.subplot(3,1,1)
sns.scatterplot(x = 'Flavanoids', y = 'OD280', data= data, hue = 'cluster',palette=['green','violet','red'])
plt.subplot(3,1,2)
sns.scatterplot(x = 'Flavanoids', y = 'Total_Phenols', data= data, hue = 'cluster',palette=['green','violet','red'])
plt.subplot(3,1,3)
sns.scatterplot(x = 'Alcohol', y = 'Proline', data= data, hue = 'cluster',palette=['green','violet','red'])
plt.show()
```

## Agglomerative Hierarchial Clustering

```python
# using the dendrogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(8,6))
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```
